```python
import yfinance as yf
import pandas as pd

# Get Tesla stock data
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")
tesla_data.reset_index(inplace=True)

# Get Tesla financials
tesla_financials = tesla.financials
tesla_revenue = tesla_financials.loc['Total Revenue'].to_frame()
tesla_revenue.reset_index(inplace=True)
tesla_revenue.columns = ['Date', 'Revenue']
tesla_revenue['Date'] = pd.to_datetime(tesla_revenue['Date']).dt.strftime('%Y-%m-%d')

print("Tesla Revenue Data:")
print(tesla_revenue.tail())

# Save the data for later use
tesla_revenue.to_csv('tesla_revenue.csv', index=False)
print("\nData saved to tesla_revenue.csv") 
```
